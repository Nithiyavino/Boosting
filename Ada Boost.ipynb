```python
import pandas as pd
```

```python
dataset=pd.read_csv("50_Startups.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['R&D Spend', 'Administration', 'Marketing Spend','State_Florida', 'State_New York']]
```

```python
independent
```

```python
dependent=dataset[['Profit']]
```

```python
dependent
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
from sklearn.ensemble import AdaBoostRegressor
regressor=AdaBoostRegressor(n_estimators=5000,learning_rate=1.0,loss='linear')
regressor=regressor.fit(x_train,y_train)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import AdaBoostRegressor
param_grid ={'n_estimators': [50, 100],'learning_rate' : [0.01,0.05,0.1,0.3,1],'loss' : ['linear', 'square', 'exponential']
}
grid = GridSearchCV(AdaBoostRegressor(), param_grid, refit = True, verbose = 3,n_jobs=-1)
grid.fit(x_train,y_train)
# fitting the model for grid
```

```python
re=grid.cv_results_
print("The R_score value for best parameter {}:".format(grid.best_params_))
```

```python
table=pd.DataFrame.from_dict(re)
```

```python
table
```
